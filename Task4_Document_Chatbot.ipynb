```python
document = {
    "python": "Python is a high-level programming language used for web development, data science, AI, and automation.",
    "machine learning": "Machine Learning is a branch of Artificial Intelligence that enables computers to learn from data.",
    "data science": "Data Science is the process of extracting insights from data using statistics and machine learning.",
    "artificial intelligence": "Artificial Intelligence is the simulation of human intelligence in machines.",
    "cloud computing": "Cloud Computing provides computing services over the internet.",
    "cyber security": "Cyber Security protects systems and networks from digital attacks.",
    "database": "A database is an organized collection of data.",
    "html": "HTML is the standard markup language for creating web pages.",
    "css": "CSS is used to style web pages.",
    "java": "Java is an object-oriented programming language."
}

while True:
    question = input("Ask a question (or type exit): ").lower()

    if question == "exit":
        print("Thank you for using the chatbot!")
        break

    found = False

    for key in document:
        if key in question:
            print(document[key])
            found = True
            break

    if found == False:
        print("Sorry, information not found in the document.")
```
